```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data=r"D:\CivicAlert\dataset\data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    workers=0,
    amp=False
)
```

```python
import shutil, glob, os

# Find the latest training run's best.pt and copy it to the project models folder
runs_dir = os.path.expanduser("~/runs/detect")
train_dirs = sorted(glob.glob(os.path.join(runs_dir, "train*")), key=os.path.getmtime)
if train_dirs:
    best_pt = os.path.join(train_dirs[-1], "weights", "best.pt")
    dest = r"D:\CivicAlert\mall-ai-system\models\best.pt"
    if os.path.exists(best_pt):
        shutil.copy2(best_pt, dest)
        print(f"Copied {best_pt} -> {dest}")
    else:
        print(f"best.pt not found at {best_pt}")
else:
    print("No training runs found")
```

```python
from ultralytics import YOLO

# After training, update the path below to your latest train run's best.pt
model = YOLO(r"D:\CivicAlert\mall-ai-system\models\best.pt")

results = model.predict(
    source=r"D:\CivicAlert\videos\input\test.jpg",
    imgsz=640,
    conf=0.25,
    show=True,
    device=0  
)
```
